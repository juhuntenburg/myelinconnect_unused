```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from vtk_rw import read_vtk
from plotting import plot_surf_stat_map, crop_img
```

```python
colors = sns.color_palette('cubehelix', 10)
```

```python
lh_mesh_file = '/scr/ilz3/myelinconnect/new_groupavg/surfs/lowres/lh_lowres_new.vtk'
lh_sulc_file = '/scr/ilz3/myelinconnect/new_groupavg/surfs/lowres/sulc/lh_sulc.npy'
lh_mask_file = '/scr/ilz3/myelinconnect/new_groupavg/masks/lh_fullmask.npy'
lh_myethick_file = '/scr/ilz3/myelinconnect/new_groupavg/myethick/myethick/smooth_6/lh_myethick_smooth_6.npy'
lh_myeratio_file = '/scr/ilz3/myelinconnect/new_groupavg/myethick/myeratio/smooth_6/lh_myeratio_smooth_6.npy'

rh_mesh_file = '/scr/ilz3/myelinconnect/new_groupavg/surfs/lowres/rh_lowres_new.vtk'
rh_sulc_file = '/scr/ilz3/myelinconnect/new_groupavg/surfs/lowres/sulc/rh_sulc.npy'
rh_mask_file = '/scr/ilz3/myelinconnect/new_groupavg/masks/rh_fullmask.npy'
rh_myethick_file = '/scr/ilz3/myelinconnect/new_groupavg/myethick/myethick/smooth_6/rh_myethick_smooth_6.npy'
rh_myeratio_file = '/scr/ilz3/myelinconnect/new_groupavg/myethick/myeratio/smooth_6/rh_myeratio_smooth_6.npy'
```

```python
lv,lf,ld = read_vtk(lh_mesh_file)
lh_mask = np.load(lh_mask_file)
lh_myethick = np.load(lh_myethick_file)
lh_myeratio = np.load(lh_myeratio_file)
lh_sulc = np.load(lh_sulc_file)

rv,rf,rd = read_vtk(rh_mesh_file)
rh_mask = np.load(rh_mask_file)
rh_myethick = np.load(rh_myethick_file)
rh_myeratio = np.load(rh_myeratio_file)
rh_sulc = np.load(rh_sulc_file)
```

```python
lh_myethick_plot = lh_myethick.copy()
lh_myethick_plot -= np.mean(np.delete(lh_myethick, lh_mask))
lh_myethick_plot[lh_mask] = 0

rh_myethick_plot = rh_myethick.copy()
rh_myethick_plot -= np.mean(np.delete(rh_myethick, rh_mask))
rh_myethick_plot[rh_mask] = 0
```

```python
lh_myeratio_plot = lh_myeratio.copy()
lh_myeratio_plot -= np.mean(np.delete(lh_myeratio, lh_mask))
lh_myeratio_plot[lh_mask] = 0

rh_myeratio_plot = rh_myeratio.copy()
rh_myeratio_plot -= np.mean(np.delete(rh_myeratio, rh_mask))
rh_myeratio_plot[rh_mask] = 0
```

### Histogramms

```python
sns.set_style('darkgrid')
sns.set_context('notebook', font_scale=2)
fig = plt.figure(figsize=(20,6))
fig.add_subplot(121)
plt.hist(np.delete(lh_myethick, lh_mask), alpha=1, color=colors[1], label='left');
plt.hist(np.delete(rh_myethick, rh_mask), alpha=0.7, color=colors[3], label='right');
plt.title('myelin thickness')
plt.legend()
fig.add_subplot(122)
plt.hist(np.delete(lh_myeratio, lh_mask), alpha=1, color=colors[0], label='left');
plt.hist(np.delete(rh_myeratio, rh_mask), alpha=0.7, color=colors[5], label='right');
plt.title('myelin ratio')
plt.legend()
```

### Myelinated thickness

```python
sns.set_style('white')
lat=plot_surf_stat_map(rv, rf, stat_map=rh_myethick_plot, bg_map=rh_sulc, bg_on_stat=True, darkness=0.5, 
                    elev=180,azim=0, figsize=(11,10.5), cmap='jet', threshold=1e-50)
sns.set_style('white')
med=plot_surf_stat_map(rv, rf, stat_map=rh_myethick_plot, bg_map=rh_sulc, bg_on_stat=True,darkness=0.5,
        elev=180,azim=180, figsize=(11,10.5), cmap='jet', threshold=1e-50)

lat_crop=crop_img(lat)
med_crop=crop_img(med)
fig=plt.figure()
fig.set_size_inches(10, 5)
ax1 = fig.add_subplot(121)
plt.imshow(lat_crop)
ax1.set_axis_off()
ax2 = fig.add_subplot(122)
plt.imshow(med_crop)
ax2.set_axis_off()
```

```python
sns.set_style('white')
lat=plot_surf_stat_map(lv, lf, stat_map=lh_myethick_plot, bg_map=lh_sulc, bg_on_stat=True, darkness=0.5, 
                    elev=180,azim=180, figsize=(11,9), cmap='jet', threshold=1e-50)

sns.set_style('white')
med=plot_surf_stat_map(lv, lf, stat_map=lh_myethick_plot, bg_map=lh_sulc, bg_on_stat=True,darkness=0.5,
        elev=180,azim=0, figsize=(11,9), cmap='jet', threshold=1e-50)
    
lat_crop=crop_img(lat)
med_crop=crop_img(med)
fig=plt.figure()
fig.set_size_inches(10, 5)
ax1 = fig.add_subplot(121)
plt.imshow(med_crop)
ax1.set_axis_off()
ax2 = fig.add_subplot(122)
plt.imshow(lat_crop)
ax2.set_axis_off()
```
